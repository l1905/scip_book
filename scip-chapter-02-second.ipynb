#### 强健设计的语言层次

分层设计这一概念说的是， 一个复杂的系统应该通过一系列的层次构造出来， 为了描述这些层次， 需要使用一系列的语言， 构造各个层次的方式， 就是
设法组合起作为这一层次中部件的各种基本元素， 面对构造出的部件又可以作为另一个层次的基本元素。 在分层设计中， 每个层次上所用的语言提供了一些
基本元素， 组合手段， 还有对该层次中适当细节做抽象的手段。在复杂系统的功能中， 广泛使用这种分层设计方法。



### 符号数据

到目前位置， 我们已经使用过的所有复合数据， 最终都是从数值触发构造出来的。 在这里， 我们要扩充

所用语言的表述能力， 引进将任意符号作为数据的功能


#### 2.3.1 引号


我们如果能构造出采用符号的复合数据， 我们就可以构造下面这些类的表

(a b c d)
(23 45 17)

为了能够操作这些符号， 我们的序言里需要一种新的元素， 为数据对象加`引号`的能力， 假定我们希望构造出表(a, b) 当然不能用(list a b)完成这件事， 因为这一表达式将要构造出的是a和b的值的表， 而不是这两个符号本身的表。 在自然语言的环境中， 这种情况也是周知的， 在哪里的单词和橘子可能看作语义实体， 也可以看作是字符的序列，（语法实体),  在自然语言里， 常见的方式就是用引号表明一个词或者一个句子应该作为问在看待， 将它们直接作为字符的序列， 例如， "John"的第一个字母显然是"J", 如果我们对某人说， “大声说出你的名字“，此时希望听到的是那个人的名字， 如果说“大声说 ‘你的名字’”， 此时希望听到的就是词组 “你的名字”， 请注意， 我们在这里不得不用嵌套的引号去描述别人应该说的东西。

我们可以按照同样的方式， 将表和符号标记为应该作为数据对象看待， 而不是作为应该求值的表达式， 然而， 这里所用的引号形式与自然语言中的不同， 我们只在被引对象的前面放一个引号 (按照习惯， 这里用单引号 ），  在Scheme里可以不写结束引号， 因为这里已经靠空白和括号将对象分隔开， 一个单引号的音译就是饮用下一个对象。


现在我们就可以区分符号和它们的值了：

```
(define a 1)

(define b 2)

(list a b)

(list 'a 'b)

(list 'a b)



```

引号也可以用于复合对象， 其中采用的是表的方便的输出表示方式：


```
(car '(a b c))

a

(cdr '(a b c))

```


记住这些之后， 我们就可以通过求值 '() 得到空表， 这样可以丢掉变量nil了


为了能够对符号做各种操作， 我们还需要用应一个基本过程  `eq?`  这个过程以两个符号作为参数， 检查它们是否为同样的符号， 利用  `eq?` 可以实现一个称为memq的有用过程， 它们以一个符号和一个表为参数， 如果这个符号不包含在这个表里， memq 就返回假， 否则返回该表的由这个符号的第一次出现的那个子表


```

(define (memq item x)
  (cond ((null? x) false)
        ((eq? item (car x)) x)
        (else (memq item (cdr x)))
  )
)

```


举例来说， 表达式

```
（memq `apple `(pear banana prune))
```

的值是假， 而表达式

```
(memq `apple `(x (apple sauce y apple pear))
```

的值(apple pear)
