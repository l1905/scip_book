## 读书笔记

### 程序设计的基本元素

1. **基本表达形式**， 用于表达语言所关心的最简单的个体
2. **组合的方法**, 通过他们可以从笔记哦啊简单的东西出发构造出符合的元素
3. **抽象的方法**, 通过他们可以为符合对象命名， 并将他们当作单元去操作

程序设计中，我们需要处理两类要素： 过程 和 数据

语言使用`前缀表示`： 将运算符放在所有运算对象的左边

不会出现歧义的原因： 运算符总是最左边的元素， 而整个表达式的范围由括号界定

 命名和环境： Lisp Scheme中， 给事务命名通过define(定义）的方式完成
 
 `define`是该语言中最简单的抽象方法， 它允许我们用一个简单的名字去引用一个组合运算结果， 一般而言， 计算得到的对象完全可以具有非常复杂的结构，  如果每次需要使用他们时， 都必须记住并重复的写出他们的细节， 那是极端不方便， 实际上构造一个复杂的程序， 也就是去一步步的创建出越来越复杂的计算性对象， 因此我们可以通过一些列的交互动作， 逐渐创建所需要的名字-对象关联， 这种特征鼓励人们采用**递增**的方式去开发和调试程序， 在很大程度上，这一情况也出于另外一个事实， 那就是，一个Lisp程序通常总是由一大批相对简单的过程组成

 **求值模型** 
 
 正则序求值： 完成展开 后规约的形式
 
 应用序求值： 先求值参数而后应用 (**Lisp采用**)
 
 条件(cond)表达式的一般性形式为:
 ```
 (cond (<p1> <e1>)
       (<p2> <e2>)
       (<p3> <e3>)
       ...
       (<pn> <en>))
 ```
 
 条件表达式的求值方式为如下： 首先求值谓词(predication)<p1>, 如果它的值是false， 那么就去求值<p2>, 如果<p2>的值是false
 就去求值<p3>, 这一过程将继续下去， 直到发现某一个谓词的值为真 则停止。此时解释器就返回相对应自居的**序列表达式**<e>的值，
 以这个值作为整个表达式的值， 如果无法找到值为真的<p>, cond的值就没有定义
    
 绝对值写法1
 ```
 (define (abs x)
     (cond ((> x 0) x)
           ((= x 0) 0)
           ((< x 0) (- x))
      )
  )
 ```
 绝对值写法2
 ```
 (define (abs x)
     (cond ((< x 0) (- x))
           (else x)
      )
  )
 ```
用自然语言来说， 就是 “如果X小于0就返回-x, 否则就返回x", else 是一个特殊符号， 可以用在
cond的最后一个字句中<e>的值
    
绝对值写法3
```
(define (abs x)
    (if (< x 0)
        (- x)
        x
     )
 )
```

这里采用的是特殊形式if, 它是条件表达式的一种受限形式， 适用于情况分析中只有两种情况的需要， if表达式的一般形式是


(if < predicate> < consequent> < alternative>)
    

除了一些**基本谓词**如 < , = 和 > , 还有一些逻辑复合运算符， 利用他们可以构造各种复合谓词， 最常用的三个复合运算符是

* (and < e1> ..... < en>)

解释器将从左到右一个一个的求值< e> , 如果某一个< e>  求值 得到假， 则 AND求值为假， 后面的那些< e> 不再求值， 如果所有前面的< e> 求为值， 
这一 AND 表达式的值就是最后那个 < e> 值    

* (or < e1> .....  < en>)

* (not < e>)

** 注意，and和or都是特殊形式而不是普通的过程， 否则他们的子表达式不一定都求值， not 则是一个普通的过程 **


### 过程作为黑箱抽象

我们无需关注这个过程是如何计算出它的结果， 只需要注意它能计算出 的事实， 关于计算过程的细节， 隐藏不提， 可以推迟到后来考虑

由此可见， 一个过程定义应该能隐藏起一些细节， 这将使过程的使用者不必自己去写这些过程， 而是从其他程序员那里作为一个黑箱而接受它， 用户在使用一个过程时， 不需要去弄清楚它时如何实现的

### 局部名

过程用户必须关心 过程里面的形式参数的名字

过程的形式参数在过程里扮演一种非常特殊的角色， 在这里， 形式参数的具体名字是什么， 这里完全没有关系， 这样的名字叫做**约束变量**, 因此我们可以说， 一个过程的定义**约束**了它的所有形式参数， 如果在一个完整的过程定义里将某个约束变量统一换名， 在者一过程定义的意义将不会左任何改变， 如果一个变量不是被约束的， 我们就称它为**自由**的， 一个名字的定义被约束与那一集表达式称作这个名字的作用与， 在一个过程定义里， 被声明为这个过程的形式参数的哪些约束变量， 就以这个过程的体作为它们的作用域。

### 内部定义和块结构
   嵌套的定义称为`块结构`, 它是最简单的名字包装问题的一种正确解决方式， 实际上， 除了可以将所用的辅助过程定义放到内部， 我们可以简化他们， 我们可以将其作为内部定义的自由变量， 这种方式，称为`词法作用域`
   
### 过程与他们所产生的计算

一个过程也就是一种模式， 它描述里一个计算过程的局部演化方式， 描述了这一计算过程中的每个步骤是怎么基于前面的步骤建立起来的， 
   
`线性迭代计算过程` 其状态可以用固定数目`状态变量`描述的计算过程， 而于此同时， 又存在一套固定的规则， 描述了计算过程从一个状态到下一个状态转换时， 这些变量的更新方式， (可能有一个)结束检测， 它描述这一计算过程应该终止的条件， 在计算n!时， 随着计算步骤随着n线性增长， 这种过程称为

在做迭代与递归之间的比较时， 我们必须当心， 不要搞混了`递归计算过程`的概念和`递归过程`的概念， 当我们说一个过程时递归的时候， 论述的时一个语法形式上的事实， 说明这个过程的定义中(直接或者间接的）引用了该过程的本身。
在说某一计算过程具有某种模式时（例如，线性递归), 我们说的时这一计算过程的**进展方式**, 而不是相应过程书写上的语法形式， 当我们说某个递归过程 将产生一个迭代的计算过程时， 可能会让人不舒服， 然而这一计算过程确实是迭代的， 因为它的状态能由其中的三个状态变量完全刻画， 解释器在执行这一种过程时， 只需要保持三个变量的轨迹足够

#### 1.2.3 增长的阶

不同的计算过程在消耗计算资源的速率上， 可能存在着巨大的差异， 描述这种差异的一种方便方式就是用**增长的阶**的计法， 以便我们理解在输入变大时， 某一计算过程中所需资源的粗略度量情况

增长的阶为我们提供了对计算过程行为的一种很粗略的描述，

#### 1.2.4 求幂

现在考虑对一个给定的数计算乘幂的问题，我们希望这一过程的参数是一个基数b和一个正整数的指数n, 过程计算出 
$$ 
\begin{align}
b^n &= b * b^{n-1} \tag{1}\\
b^0 &= 1 \tag{2}\\
\end{align}
$$

可以直接翻译为递归模式
```
(define (expt b n)
    (if (= n 0)
        1
        (* b (expt b (- n 1)))
    )
)
```
这是一个线性的递归计算过程， 需要$\Theta(n)$步 和$\Theta(n)$空间， 就像阶乘一样， 我们很容易将其形式化为一个等价的线性迭代：

```
(define (expt b n)
    (expt-iter b n 1)
)
(define (expt-iter b counter product)
    (if (= counter 0)
        product
        (expt-iter b (- counter 1) (* b product))
    )
)
```
这一版本需要$\Theta(n)$步和$\Theta(1)$空间

我们可以通过连续求平方， 以更少的步骤完成乘幂计算。 例如， 不是采用下面的方式计算$b^8$
$$ 
b \cdot ( b \cdot ( b \cdot ( b \cdot ( b \cdot ( b \cdot ( b \cdot ( b \cdot))))))))
$$
而是用三次乘法算出来:
$$ 
\begin{align}
b^2 &= b * b \tag{1}\\
b^4 &= b^2 \cdot b^2 \tag{2}\\
b^8 &= b^4 \cdot b^4 \tag{3}\\
\end{align}
$$
这一方法对于指数为2的乘幂都可以用， 如果采用下面的规则， 我们就可以借助于连续求平方， 去完成一般的乘幂计算:
$$
\begin{align}
b^n &= (b^{n/2})^2 \tag{1}\\
b^n &= b \cdot b^{n-1} \tag{2}\\
\end{align}
$$

我们可以定义如下的过程
(define (fast-expt b n)
    (cond ((= n 0) 1)
          ((even n) (square (fast-expt b (/ n 2)))
          (else (* b (fast-expt b (- n 1))))
    )
    
)
检测一个整数是否偶数的谓词可以基于基本过程remainder定义
(define (even n)
    (= (remainder n 2) 0)
)

存在着一种以对数步数求出斐波那契数的巧妙算法， 请回忆1.2.2节fib-iter计算过程中状态变量a和b的变换规则， a $\gets$ = a + b 和 b $\gets$ a, 现在将这种变换称为T变换， 通过观察可以发现， 从1和0开始将T反复应用n次， 将产生出一对数Fib(n+1) 和Fib(n). 换句话说， 斐波那契数可以通过将$T^n$(变换T的n次方)应用于对偶(1, 0)而产生出来。 现在将T看作是变换族Tpa中p=0 且q=1 的特殊情况， 其中Tpq是对于对偶(a, b) 按照a$\gets$bq + aq + ap 和 b$\gets$ bp + aq的规则变换， 请证明， 如果我们应用变换T

```
(define (fib n)
    (fib-iter 1 0 0 1 n)
)

(define (fib-iter a b p q count)
    (cond ((= count 0) b)
          ((even count)
                (fib-iter 
                        a 
                        b
                        (+ (* p q) (* q q)) ; compute 'p'
                        (+ (* 2 p q) (* q q))
                        (/ count 2)
                )
          )
          (else
                (fib-iter
                    (+ (* b q) (* a q) (* a p))
                    (+ (* b p) (* a q))
                    p
                    q
                    (- count 1)
                )
          )
    )
)

;是否是偶数
(define (even n2)
    (= (remainder n2 2) 0)
)
```

#### 1.2.5 最大公约数

两个整数a 和b的最大公约数(GCD)定义为能除尽这两个数的那个最大整数， 例如， 16和28的 GCD就是4， 在第2章里， 当我们要去研究有理数算术的实现时， 就会需要GCD， 以便能把有理数约化到最简形式(要将有理数约化到最简形式， 我们必须将其分母和分子同时除掉他们的GCD， 例如16/28 将约简为4/7, 找出两个整数GCD的一种方式是对它们做因数分解， 并从中找出公共因子， 但存在一个更高效的著名算法。

这一算法的思想基于下面的观察： 如果r是a除以b的余数， 那么a和b的公约数正好也是b和r的公约数， 因此我们可以借助于等式：
$$ GCD(a, b) = GCD(b, r)$$
这就把一个GCD的计算问题连续的归约到越来越小的整数对的GCD的计算问题。 例如
$$
\begin{align}
GCD(206, 40) &= GCD(40, 6) \tag{1}\\
 &= GCD(6, 4) \tag{2}\\
 &= GCD(4, 2) \tag{3}\\
 &= GCD(2, 0) \tag{4}\\
 &= 2 \tag{5}\\
\end{align}
$$

将GCD(206, 40)归约到GCD(2, 0), 最终得到2, 可以证明， 从任意两个正整数开始， 反复执行这种规约， 最终将产生出一个数对， 其中的第二个数是0， **此时的GCD就是另一个数**。这计算GCD的方法称为**欧几里得算法**
```
(define (gcd a b)
    (if (= b 0)
        a
        (gcd b (remainder a b))
    )
)
```

**Lame定理**：如果欧几里得算法需要用K步计算出一对整数的GCD, 那么这对数中较小的那个数， 必然大于或者等于第K个斐波那契数

我们可以利用这一定理， 作出欧几里得算法的增长阶估计。 令n是作为过程输入的两个数中较小的那个， 如果计算过程需要K步， 那么我们就一定有
$ n >= Fib(k) ==a^{k}/ \sqrt{5}$, 这样步数的增长就是n的对数 ， 这样算法的增长阶就是$\Theta(logn)$

#### 1.2.6 素数检测

1. 寻找因子， 代码例子实现
2. 费马检测

$\theta(log n)$ 的素数检查基于数论中著名的费马小定理的结果

**费马小定理**：如果n是一个素数， a是小于n的任意正整数， 那么a的n次方与a模n同余。(两个数称为是模n同余，如果它们除以n的余数相同。数a除以n的余数称为a取模n的余数， 或称为a取模n）

如果n不是素数， 那么，一般而言，大部分的 a<n 都满足上面的关系， 这就引出了下面检查素数的算法： 对于给定的整数n， 随机任取一个 a<n 并计算出$a^n$取模n的余数。如果得到的结果不等于a， 那么n就肯定不是素数。 如果它就是a， 那么n是素数的机会就很大。 现在再另取一个随机的a并采用同样的方式检查， 如果它满足上述等式，那我们就对n是素数有更大的信心， 通过检查越来越多的a值， 我们就可以不断增加对有关结果的信心，这一算法称为费马检查
